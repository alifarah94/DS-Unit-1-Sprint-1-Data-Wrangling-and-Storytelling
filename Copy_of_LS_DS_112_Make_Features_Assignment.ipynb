```python
import pandas as pd

df = pd.read_csv('LoanStats_2018Q4.csv',header=1, skipfooter=2) # the first row becomes header and removes the last 2 rows
df # check change
```

```python
df.isnull().sum() # I was able to remove a lot of the null values from each column with skipfooter. I assume that the other columns with a lot null values have private data and data hasnt been given.
```

```python
# i had issues looping through the dataset due to google colab limited ram so only loops through the first 100 elements on the list. change the loop number when testing

def month_digit(x):
  new_list = []
  new_number = ""
  for i in range(1000):
    new_number = x[i].strip() # I looked through the data and noticed that every element in the column has 2 leading white spaces so I used the strip functing
    new_number = int(new_number[:2]) # then take the first 2 letters, turn it into an integer and then reasign it 
    new_list.append(new_number) #append it to list
  return new_list # and then return the list
```

```python
df.term[:1000] = month_digit(df.term) # change index when testing. turns the first 1000 elements into an integer 
```

```python
df.term.value_counts() # check change. term now number
```

```python
df['loan_status_is_great'] = 0

# a for loop to go through 1000 element. checks current or fully paid, adds a 1 to the column anything else adds 0 to the column  
# did not make a function due to time. takes very long to troubleshoot
for i in range(1000):
  if df['loan_status'][i] =="Current" or "Fully Paid":
    df['loan_status_is_great'][i] = 1
  else:
    df['loan_status_is_great'][i] = 0
    
```
